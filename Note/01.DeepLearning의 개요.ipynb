#### Deep Learning
- 머신러닝의 한 종류
- 인공신경망은 인간의 뇌신경망의 신호체계를 본따서 만든 구조
- 이미지 인식, 음성 인식, 자연어 처리 등(비정형 데이터)의 다양한 분야에 활용

----
##### Deep Learning 과 Machine Learning의 차이점
- 가장 큰 차이점은 '특징량(Feature) 추출'
- 머신러닝은 사람(분석가)이 이미지 내부에 있는 특징을 지정하는 방식
- 딥러닝에서는 학습데이터에서 기계가 자동으로 특징을 추출
- Deep Learning은 스스로 학습을 하기 때문에 Machine Learning 보다 많은 데이터가 필요
- 즉, 머신러닝까지는 인간이 가르쳐야 했지만 딥러닝을 이용하면 기계가 자동으로 특징을 학습할 수 있다.

----
#### 신경망(Neural Network)
- 인간의 신경망을 본따 만든 네트워크 구조를 의미
- 입력층, 은닉층, 출력층을 지나 처리가 일어나고 최종적인 결과가 출력
- 이러한 신경망을 은닉층의 2개 이상 추가되면 깊은 신경망(Deep Neural Network)이라고 하며, 이를 활용한 기계학습을 딥러닝이라 한다.
- 딥러닝은 대량의 데이터를 입력해서 학습한다.
- 학습이란 유닛(Unit)의 접속 가중치를 조정하는 것을 의미

----
#### Deep Learning이 Machine Learning보다 우수한 분야
- Image Classification
- Image Object detection
> 발전된 분야 : 자율주행차 (이미지를 보고 차가 사람인지 아닌 지를 분류)

- Sequence data 분석 및 예측(순서가 있는 데이터)
> 발전된 분야 : 번역기(어순이 있다.)

----
#### 용어 정의

#### 손실함수(Loss Function)
- 출력값(예측값)과 정답의 차이
- 출력값과 정답이 일치할 수록 손실함수의 값은 작아진다.
- 회귀에는 평균절대오차(MAE :Mean Absolute Error)와 평균제곱오차(MSE : Mean Suared Error)
- 분류 문제에는 크로스 엔트로피(Cross Entropy)를 사용
- 매개변수(Weight, Bias)를 조절해서 손실함수의 값을 최저로 만드는 과정을 최적화(Optimization)이라 한다.
- 최적화 과정은 Optimizer를 통해 이뤄지며, Optimizer는 역전파(back propagation)과정을 수행해서 딥러닝 모댈의 매개변수를 최적화 한다.

----
#### 최적화(Optimization)
- 대표적인 방법은 경사하강법(Batch Gradient Decent)
- 반복적으로 손실함수에 대한 모델 매개변수의 미분값(기울기)를 구한 후 그 미분값의 반대방향으로 매개변수를 조절해 나가면 결국에는 최저 손실함수에 도달한다.


#### 역전파 (Back Propagation)
- Optimizer는 손실함수의 값을 최소화 하기 위해 역전파를 사용해 딥러닝 모델의 모든 매개변수를 변경한다.
- 손실함수의 값을 최소화 한다는 것은 정답과 예측값의 차이를 최소화 한다는 것이며 에러율을 최저로 줄인다는 의미

#### Optimizer
- Batch 경사 하강법
- SGD(Stochastic Gradient Descent) : 기울기가 급할 땐 step을 넓게, 기울기가 완만해지면 step을 좁게

#### 미니배치(mini batch)
- 배치 경사법은 너무 느리고 리소스도 많이 사용 / SGD는 확실히 빨리 학습하지만 모델이 최적화 안되는 경우가 많습니다.
- 두 방법의 절충안이 미니배치
- 전체 데이터를 계산해서 매개변수를 변경하는 대신 정해진 양만큼 계산해서 매개변수를 최적화 하는 방법
- 최근 딥러닝 학습에서는 대다수가 미니배치를 사용

#### 학습률(Learning Rate)
- Adagrad : 자연어처리
- Adam : 이미지

#### 딥러닝의 과대적합(Overfitting)
- 과대적합은 학습시와 검증시의 예측력의 차이가 심한 현상
- 딥러닝 모델은 학습시 매개변수가 상당히 많다는 점과 학습 횟수에 제한이 없어서 과대적합을 불러올 수 있다.

#### 드롭아웃(Drop Out)
- 매개변수 중 일정량을 학습 중간마다 무작위로 사용하지 않는 방법

#### 조기 종료(Early Stopping)
- 무조건 학습 반복 횟수(epoch)를 높이면 학습시간이 길어져서 학습데이터만 성능이 좋은 현상 발생
- 어느정도 학습률이 좋으면 사용자가 정한 반복횟수 이전에 작업을 종료
