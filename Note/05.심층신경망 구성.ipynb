#### 패션 MNIST

```python
from tensorflow import keras

(train_input, train_target), (test_input, test_target) = \
   keras.datasets.fashion_mnist.load_data()
```

```python
# train과 valid data 분리
from sklearn.model_selection import train_test_split

train_scaled = (train_input / 255.0).reshape(-1,28*28)

train_scaled, val_scaled, train_target, val_target = \
   train_test_split(
      train_scaled,
      train_target,
      test_size=0.2,
      random_state=42
   )
```

#### Layer를 추가하는 방법 -1

```python
# Layer 만들기
dense1 = keras.layers.Dense(100,activation='sigmoid', input_shape=(784,)) # 은닉층
dense2 = keras.layers.Dense(10,activation='softmax') # 출력층
```

```python
# 전체 모델의 층을 추가
model = keras.Sequential([dense1, dense2])
```

```python
model.summary()
```

```python
model1 = keras.Sequential(
   [
      keras.layers.Dense(100,activation='sigmoid', input_shape=(784,), name='hidden'),
      keras.layers.Dense(10,activation='softmax', name='output')
   ],
   name= "패션 MNIST 모델"
)
```

----
#### Layer를 추가하는 방법 - 3

```python
model = keras.Sequential()
model.add(keras.layers.Dense(100,activation='sigmoid', input_shape=(784,),name='hidden')),
model.add(keras.layers.Dense(10,activation='softmax',name='output'))
```

> 제일 많이 사용하는 방법이다. 파이썬 코드가 중간에 삽입이 가능하다.

```python
# 모델 설정 (Compile)
model.compile(
   loss="sparse_categorical_crossentropy", 
   metrics=['accuracy']                    
)

# 모델 학습 (Fit)
model.fit(
   train_scaled,   # 훈련 입력 데이터
   train_target,   # 훈련 정답 데이터
   epochs=5        
)
```

#### 입력층을 이미지 그대로 사용하기

```python
(train_input, train_target), (test_input, test_target) = \
   keras.datasets.fashion_mnist.load_data()

train_scaled = train_input / 255.0

train_scaled, val_scaled, train_target, val_target = \
   train_test_split(
      train_scaled,
      train_target,
      test_size=0.2,
      random_state=42
   )
```

```python
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28,28)))
model.add(keras.layers.Dense(100,activation='relu'))
model.add(keras.layers.Dense(10,activation='softmax'))
```

```python
# Model
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28,28)))
model.add(keras.layers.Dense(100,activation='relu'))
model.add(keras.layers.Dense(10,activation='softmax'))

# 손실함수
model.compile(
   loss="sparse_categorical_crossentropy",
   metrics=['accuracy']
)

# 학습하기
model.fit(
   train_scaled,
   train_target,
   epochs=24
)

# 테스트 데이터
test_scaled = test_input / 255.0

# 과대적합 분포 확인
print("Train :", model.evaluate(train_scaled, train_target))
print("Valid :", model.evaluate(val_scaled, val_target))
print("Test  :", model.evaluate(test_scaled, test_target))
```
